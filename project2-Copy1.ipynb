```python
pageWwGross2016 = responsewwGross2016.text
soupWwGross2016 = BeautifulSoup(pageWwGross2016)

tableWwGross2016 = soupWwGross2016.find_all('table')

#check the len of each element of above table:
for element in tableWwGross2016:
    print(len(element))
```

```python
listOfTable = []

for rowNum in range(2, len(dataJumble)):
    listOfData = dataJumble[rowNum].find_all('td')
    rowData = []
    for i in listOfData[1:]:
        rowData.append(i.text)
    rowData.append(listOfData[1].find('a').get('href'))
    listOfTable.append(rowData)

import pandas as pd

df = pd.DataFrame(listOfTable)
df.columns = ['Movie','Distributor','WW_gross','Domestic_gross','Domestic_percentage','Intr_gross','Intr_percentage','Movie_link']
df
```
