```python
from sklearn import cross_validation
from sklearn import linear_model
from sklearn.ensemble import RandomForestRegressor
```

```python
def calcPercentage(numerator, denominator):
    out = numerator/denominator*100.
    return out

dfcomb3['China_percentage'] = dfcomb3.apply(lambda row: calcPercentage(row.China,row.WW_gross), axis=1).fillna(0)
dfcomb3['Russia_percentage'] = dfcomb3.apply(lambda row: calcPercentage(row['Russia - CIS'],row.WW_gross), axis=1).fillna(0)
dfcomb3['UK_percentage'] = dfcomb3.apply(lambda row: calcPercentage(row['United Kingdom'],row.WW_gross), axis=1).fillna(0)
dfcomb3['Japan_percentage'] = dfcomb3.apply(lambda row: calcPercentage(row.Japan,row.WW_gross), axis=1).fillna(0)
```

```python
dfcomb3['Budget_percentage'] = dfcomb3.apply(lambda row: calcPercentage(row.ProdBudget,row.WW_gross), axis=1)
```

```python
# if we try to map from categorical to numerical
dist_val = list(dfcomb3.Dist.unique())
dist_index = [i for i in range(len(dist_val))]
len(dist_index)

dist_mapping = dict(zip(dist_index,dist_val))
dfcomb3['Dist'] = dfcomb3['Dist'].map(dist_mapping)
```

```python
dfcomb3.drop('Dist', axis=1)
dfcomb4 = dfcomb3.drop('Dist', axis=1).dropna(axis=0, how='any', inplace=False)
len(dfcomb4)
dfcomb4
```

```python
'''
trainCols = ['Dist','Distributor','Domestic_percentage','Intr_percentage','MpaaRating','UsDist',
             'UsRelease','WW_gross','movie', 'updatedGenre', 'China_percentage','Russia_percentage',
             'UK_percentage', 'Japan_percentage','Budget_percentage']

'''
trainCols = ['Domestic_percentage','Intr_percentage','WW_gross', 'China_percentage','Russia_percentage',
             'UK_percentage', 'Japan_percentage','Budget_percentage']

X = dfcomb4[trainCols]
y = dfcomb4['WW_gross']

X
```

```python
X_train, X_test, y_train, y_test = cross_validation.train_test_split(X, y, test_size=0.3)
```

```python
rfmodel = RandomForestRegressor(n_estimators=500, max_features=4)
```

```python
rfmodel.fit(X_train,y_train)
```

```python
rfmodel.score(X_test, y_test)
```
